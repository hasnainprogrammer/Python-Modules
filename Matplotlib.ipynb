# Matplotlib Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

# The plot() method:-

The plot() method is used to plot a diagram between x and y axis.

```python
x = np.array([1,8])
y = np.array([10,50])
plt.plot(x,y)
```

if you do not want the line just want the point then just add string 0 as another parameter.

```python
plt.plot(x,y,'o')
```

You can also used a marker depending on your choice,just pass the keyword marker as argument to the plot function.

```python
plt.plot(x,y,marker='*')
```

you can change the border style as well as the color of the line.

: => Means dotted border

r => red color

```python
plt.plot(x,y,'o:r')
```

You can change the size and the color of the marker.

ms => Marker size

mec => Marker edge color

mfc => Marker face color


```python
plt.plot(x,y,marker='o',ms=20) # Marker size
```

```python
plt.plot(x,y,marker='o',ms=20,mec='r') # Marker Edge color
```

```python
plt.plot(x,y,marker='o',ms=20,mfc='g') # Marker Face color
```

```python
plt.plot(x,y,marker='o',ms=20,mec='hotpink',mfc='hotpink') # Marker edge color as well as Marker face color.
```

```python
plt.plot(x,y,linestyle='dashed',color='hotpink') # Line Style and line color.
```

**---> You can plot as many points as you want.**

```python
plt.plot(x)
plt.plot(y)
```

# Creating a labels

```python
plt.title('My Graph') 
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
```

```python
# You can also set the properties of font such as font family,font color or font size etc.
font1 = {'family':'Arial','color':'red','size':30}
font2 = {'family':'serif','color':'green','size':20}

plt.title('My New Graph',fontdict=font1,loc='left') # You can use the loc method to align the title as you want.
plt.xlabel('X-Axis',fontdict=font2)
plt.ylabel('Y-Axis',fontdict=font2)
plt.plot(x,y)
```

**Grid Lines**

```python
plt.grid() # the grid function add a grid line to the graph.
plt.plot(x,y)
```

```python
# Specify which grid line should display,E.g:- X or Y axis.
plt.grid(axis='x') # x-axis line
plt.plot(x,y)
```

```python
plt.grid(axis='y') # y-axis line
plt.plot(x,y)
```

```python
plt.grid(color='green',linestyle='dotted') # you can also specify the line properties such as color or linestyle etc.
plt.plot(x,y)
```

# Scatter plot

```python
plt.scatter(x,y)
```

```python
# Comparing two plots
a = np.array([5,4,3])
b = np.array([8,7,6])
plt.scatter(x,y)
plt.scatter(a,b)
```

```python
# Chaging the color
plt.scatter(x,y,c='purple')
plt.scatter(a,b,c='cyan')
```

```python
# You can also specify different color for every point using only c not with color word by passing a list of different colors. 
colors = ['red','green','blue']
plt.scatter(x,y)
plt.scatter(a,b,c=colors)
```

```python
# changing the size of the dot
plt.scatter(x,y,s=1000,c='hotpink')
plt.scatter(a,b,s=500,c='orange')
```

```python
# You can also change the size of each dot.
size1 = [100,200]
size2 = [500,700,1000]
plt.scatter(x,y,s=size1)
plt.scatter(a,b,s=size2)
```

```python
# Changing the transperancy of the color of dot.
size1 = [100,200]
size2 = [500,700,1000]
plt.scatter(x,y,s=size1,alpha=0.5)
plt.scatter(a,b,s=size2,alpha=0.5)
```
